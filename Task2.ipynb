```python
# Importing libraries
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('Au_nanoparticle_dataset (1).csv')
df.head(15)
```

```python
filtered_data = df.drop(columns=['N_total', 'N_bulk', 'N_surface', 'R_avg'])
```

```python
print(filtered_data.head(20))
```

```python
selected_features = df[['N_total', 'N_bulk', 'N_surface', 'R_avg']]

mean_values = selected_features.mean()
std_values = selected_features.std()
quartiles = selected_features.quantile([0.25, 0.5, 0.75])

print("Mean values:\n", mean_values)
print("Standard deviation:\n", std_values)
print("Quartile values:\n", quartiles)
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 4, figsize=(20, 5))
selected_features = ['N_total', 'N_bulk', 'N_surface', 'R_avg']

for i, feature in enumerate(selected_features):
    axes[i].hist(df[feature], bins=30, edgecolor='black')
    axes[i].set_title(feature)
    axes[i].set_xlabel('Value')
    axes[i].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
g = sns.pairplot(filtered_data)
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
plt.show()
```

```python
g = sns.PairGrid(filtered_data)
g.map_upper(sns.histplot)  # bivariate histogram
g.map_diag(sns.histplot, kde=True)  # histogram with KDE on diagonal
g.map_lower(sns.kdeplot)  # bivariate KDE plot on lower half
plt.show()
```
