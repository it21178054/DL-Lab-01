```python
plt.hist(np.random.rand(100000), density=True, bins=100, histtype="step", color="blue", label="rand")
plt.hist(np.random.rand(100000), density=True, bins=100, histtype="step", color="red", label="uniform")
plt.hist(np.random.randn(100000), density=True, bins=100, histtype="step", color="green", label="normal")
plt.axis([-2.5, 2.5, 0, 1.1])
plt.legend(loc = "upper left")
plt.title("Random distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

# Create meshgrid
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

# Plot 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_title("Z = X^2 + Y^2")
ax.set_xlabel("X axis")
ax.set_ylabel("Y axis")
ax.set_zlabel("Z axis")
plt.show()
```

```python
import pandas as pd
import requests
import seaborn as sns
import matplotlib.pyplot as plt

# Download the dataset using requests
url = 'https://elitedatascience.com/wp-content/uploads/2022/07/Pokemon.csv'
response = requests.get(url)

# Save the content to a CSV file
with open('Pokemon.csv', 'wb') as file:
    file.write(response.content)

# Read the dataset
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')
print(df.head())

# Calculate Pearson and Spearman correlation coefficients for selected features
features = ["HP", "Attack", "Defense", "Sp. Atk", "Sp. Def", "Speed"]
data_subset = df[features]

# Calculate correlations
pearson_corr = data_subset.corr(method='pearson')
spearman_corr = data_subset.corr(method='spearman')

# Visualize using heatmaps
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm')
plt.title("Pearson Correlation")

plt.subplot(1, 2, 2)
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm')
plt.title("Spearman Correlation")
plt.show()
```
